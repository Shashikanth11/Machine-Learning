```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('adults_2550_male.csv')

# Select relevant features for clustering
features = ['amt', 'category', 'merchant', 'amt_log', 'merchant_freq', 'trans_count', 'total_amt']

# Standardize text data to lowercase to ensure uniformity in categorical data
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.lower()

# Encode categorical features
le = LabelEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Drop duplicates from the dataset
df.drop_duplicates(inplace=True)

# Handle missing values by removing rows with NaN values in selected features
df.dropna(subset=features, inplace=True)

# Extract features for clustering
X = df[features]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values to visualize the Elbow point
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Apply KMeans clustering with the selected number of clusters (e.g., 2 for high and low spenders)
optimal_clusters = 2  # Adjust this based on the Elbow plot
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Analyze the clusters by looking at average feature values for each cluster
print("Cluster Centers:")
print(df.groupby('cluster')[features].mean())

# Check the number of customers in each cluster
print("\nCluster Sizes:")
print(df['cluster'].value_counts())

# Assign descriptive labels for marketing insights
df['spending_category'] = df['cluster'].apply(lambda x: 'High Spender' if x == 0 else 'Low Spender')

# Display a sample of the clustered data with spending labels
print("\nSample Clustered Data:")
print(df[['amt', 'total_amt', 'spending_category']].head())
```

```python
from sklearn.metrics import silhouette_score
sil_score_kmeans = silhouette_score(X_scaled, df['cluster'])
print(f"Silhouette Score for K-Means: {sil_score_kmeans}")
#
```
